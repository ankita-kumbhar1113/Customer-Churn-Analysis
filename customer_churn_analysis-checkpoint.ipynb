```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
#replacing blanks with 0 
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
#converted 0 and 1 value to yes/no of seniorcitizen
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title('Churn Count Distribution ')
plt.show()
```

```python
ax = sns.countplot(x="gender", data = df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure", data=df,hue = "Churn", bins = 72)
plt.title("Churn by Tenure")
plt.show()
```

```python
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("count of customers by contract")
plt.show()
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies' ]
fig,axes = plt.subplots(3,3, figsize=(18,15))
axes = axes.flatten()
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], palette='Set1', hue = "Churn")
    axes[i].set_title(f'Count of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn of customers using Paymentmethod")
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df, hue='Churn')
plt.title("Tenure VS Churn")
plt.show()
```
